```python
# import packages
import pandas as pd
import plotly.express as px
```

```python
# load data
gdp_df = pd.read_csv("ds4200_project_co2_data.csv")
```

```python
# getting rid of NA values
filter_gdp_df = gdp_df.dropna(subset=["iso_code", "gdp"])

# creating chlorpleth figure
fig = px.choropleth(
    filter_gdp_df,
    locations="iso_code", # uses iso_code because of plotly internal database
    color="gdp",
    hover_name="country",
    animation_frame="year",
    title="GDP of Countries Over Time",
    color_continuous_scale="Viridis",
    labels={"gdp": "GDP (USD)"}
)

# displaying figure
fig.show()
```
